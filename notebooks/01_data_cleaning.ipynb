## Transaction Cleaning and Feature Creation

Cancelled transactions and invalid records were removed from the dataset. A Sales feature was then created using Quantity and UnitPrice for subsequent revenue and customer value analysis.

```python
import pandas as pd

df = pd.read_csv("../data/cleaned/cleaned_retail_day2.csv")

print(df.shape)
```

```python
# Check cancelled invoices
df["InvoiceNo"].astype(str).str.startswith("C").sum()
print(df.shape)
```

```python
# Remove cancelled invoices
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
print(df.shape)
```

```python
# Remove negative quantities
df = df[df["Quantity"] > 0]
```

```python
# Remove negative prices
df = df[df["UnitPrice"] > 0]
```

```python
# Create Sales column
df["Sales"] = df["Quantity"] * df["UnitPrice"]
```

## Conclusion

The Online Retail dataset was successfully cleaned and prepared for analysis. Missing customer identifiers, duplicate records, cancelled transactions, and invalid values were removed.
